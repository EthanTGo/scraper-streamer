```python
'''
I will use this to demonstrate how the streaming pipeline will work for the project
'''
import numpy as np
import pandas as pd
import datetime

# Install the following with pip if haven't
import tweepy

# config.py that contains the consumer key and access token
from config import *
```

```python
auth = tweepy.OAuthHandler(TWITTER_CONSUMER_KEY, TWITTER_CONSUMER_SECRET)
auth.set_access_token(TWITTER_ACCESS_TOKEN, TWITTER_ACCESS_TOKEN_SECRET)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
# Setting up the date
today = datetime.date.today()
yesterday = today - datetime.timedelta(days = 1)
```

```python
tweets_list = tweepy.Cursor(api.search_tweets, q="#Covid-19 since:" + str(yesterday)+ " until:" + str(today),
                            tweet_mode='extended', lang='en').items()
```

```python
output = []

num_tweets = 0
for tweet in tweets_list:
    if(num_tweets == 100):
        break
    text = tweet._json["full_text"]
    print(text)
    favourite_count = tweet.favorite_count
    retweet_count = tweet.retweet_count
    created_at = tweet.created_at
    
    line = {'text' : text, 'favourite_count' : favourite_count, 'retweet_count' : retweet_count, 'created_at' : created_at}
    output.append(line)
    num_tweets += 1
```

```python
df = pd.DataFrame(output)
df.to_csv('output.csv')
```
